# Using SciPy to Calculate Bessel Functions

This example calculates the values for Bessel functions of the first kind in a given range. The values should agree with Bergman 2011, Table B.4, on page 1017.

## bessel.py

```python
"""
Example of using SciPy for bessel functions of the first kind. The calculated
values are printed to the console.

Reference:
Bergman, Lavine, Incropera, Dewitt 2011 from Ch. 5, pg.299-304, pg.1017
"""

# Modules
#------------------------------------------------------------------------------

import numpy as np
import scipy.special as sp

# Bessel Function Example
#------------------------------------------------------------------------------

# header for the data printed in the console
print('step\t J0\t J1')

# calculate and print the Bessel function values in a given range
# the values should agree with Bergman 2011, Table B.4, pg.1017
for i in np.arange(0, 2.5, 0.1):
    j0 = sp.j0(i)
    j1 = sp.j1(i)
    print('{:.4f}\t {:.4f}\t {:.4f}'.format(i, j0, j1))
    
```

# Positive Roots of the Transcendental Equations

Plots the $\zeta, Bi$ (transcendental) equations for a sphere, cylinder, and slab. Note that $\zeta_n$ are positive roots of the function. The Biot number is given by $Bi = h r_o / k$. Before the plots, the first 5 roots are printed to the console for reference.

$$\text{slab:} \quad \zeta_n\, tan(\zeta_n) = Bi$$  
$$\text{cylinder:} \quad \zeta_n \frac{J_1(\zeta_n)}{J_0(\zeta_n)} = Bi$$  
$$\text{sphere:} \quad 1 - \zeta_n\, cot(\zeta_n) = Bi$$  

The transcendental equations are written as functions in the `funcZeta.py` file and the calculations to determine the positive roots are performed in the `funcRoots.py` file. See the comments in each function file for more details.

## roots.py

```python
# create inline plots for this notebook
%matplotlib inline
```

```python
"""
Plot the zeta, Bi equation and its positive roots. For use in the analytical
transient heat conduction of a 1D sphere, cylinder, or slab.

Functions:
funcRoots.py returns the positive roots of the zeta, Bi equation
funcZeta.py functions for zeta, Bi equation for sphere, cylinder, and slab

References: 
1) Recktenwald 2006
2) Bergman, Lavine, Incropera, Dewitt 2011 from Ch. 5, pg.299-304
"""

# Modules and Functions
#------------------------------------------------------------------------------
 
import numpy as np
import matplotlib.pyplot as py

from funcRoots import roots
from funcZeta import funcZetaSph, funcZetaCyl, funcZetaSlab

py.close('all')
py.rcParams['figure.figsize'] = (8.0, 6.0)

# Parameters
#------------------------------------------------------------------------------

Bi = 5  # Biot number

# range of z-values from 0 to 15 as 100 samples
# the range at which the zeta, Bi equation (function) will be evaluated 
z = np.linspace(0, 15, num=100)

# Sphere Positive Roots and Zeta, Bi Equation
#------------------------------------------------------------------------------

b = 2   # shape factor where 2 sphere, 1 cylinder, 0 slab

rts_sph = roots(z, b, Bi)   # positive roots of the zeta, Bi equation
print('roots of sphere zeta, Bi equation \n', rts_sph)

fz_sph = funcZetaSph(z, Bi) # evaluate zeta, Bi sphere equation at z-values 

# Cylinder Positive Roots and Zeta, Bi Equation
#------------------------------------------------------------------------------

b = 1   # shape factor where 2 sphere, 1 cylinder, 0 slab

rts_cyl = roots(z, b, Bi)   # positive roots of the zeta, Bi equation
print('roots of cylinder zeta, Bi equation \n', rts_cyl)

fz_cyl = funcZetaCyl(z, Bi) # evaluate zeta, Bi cylinder equation at z-values 

# Slab Positive Roots and Zeta, Bi Equation
#------------------------------------------------------------------------------

b = 0   # shape factor where 2 sphere, 1 cylinder, 0 slab

rts_slab = roots(z, b, Bi)   # positive roots of the zeta, Bi equation
print('roots of slab zeta, Bi equation \n', rts_slab)

fz_slab = funcZetaSlab(z, Bi) # evaluate zeta, Bi slab equation at z-values

# Plot Each Function and its Positive Roots
#------------------------------------------------------------------------------
    
py.figure(1)
py.plot(z, fz_sph, lw=2)
py.scatter(rts_sph, np.zeros(len(rts_sph)), c='r', s=60, edgecolor='none')
py.axhline(y=0, c='k', ls='--')
py.ylabel('f ($\zeta$)')
py.xlabel('$\zeta$')
py.title('Sphere at Bi={}'.format(Bi))
py.ylim([-20, 20])
py.xlim(0)
py.grid()
py.show()

py.figure(2)
py.plot(z, fz_cyl, lw=2)
py.scatter(rts_cyl, np.zeros(len(rts_cyl)), c='r', s=60, edgecolor='none')
py.axhline(y=0, c='k', ls='--')
py.ylabel('f ($\zeta$)')
py.xlabel('$\zeta$')
py.title('Cylinder at Bi={}'.format(Bi))
py.ylim([-20, 20])
py.xlim(0)
py.grid()
py.show()

py.figure(3)
py.plot(z, fz_slab, lw=2)
py.scatter(rts_slab, np.zeros(len(rts_slab)), c='r', s=60, edgecolor='none')
py.axhline(y=0, c='k', ls='--')
py.ylabel('f ($\zeta$)')
py.xlabel('$\zeta$')
py.title('Slab at Bi={}'.format(Bi))
py.ylim([-20, 20])
py.xlim(0)
py.grid()
py.show()
```
